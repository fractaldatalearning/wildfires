In order to tell where the DOY comes each year that lands at the point where 10% of fires have started, it will be helpful to have a column indicating an order or count of fires. 

```python
%%sql
ALTER TABLE fires    
    ADD COLUMN count_this_year INTEGER
```

```python
%%sql
-- first just take a look at how I want this column to look w/ a query, before filling it in.
SELECT OBJECTID, DISCOVERY_DATE, 
    ROW_NUMBER() OVER(PARTITION BY FIRE_YEAR ORDER BY DISCOVERY_DOY, discovery_time) 
    AS count_this_year
FROM fires
ORDER BY FIRE_YEAR
LIMIT 200
```

```python
%%sql 
-- Build the column in a temp table so it can be added to the original table
CREATE TEMP TABLE count_table AS
    WITH ct_cte AS
        (SELECT OBJECTID, DISCOVERY_DATE, 
             ROW_NUMBER() OVER(PARTITION BY FIRE_YEAR ORDER BY DISCOVERY_DOY, discovery_time) 
               AS count_this_year
         FROM fires)
SELECT * FROM  ct_cte
```

```python
%%sql
-- Check temp table. 
SELECT * 
FROM count_table 
LIMIT 200
```

```python
%%sql
-- Fill in column in original table with vaules from temp table
UPDATE fires AS f
    SET count_this_year = c.count_this_year
    FROM count_table as c
    WHERE c.OBJECTID = f.OBJECTID
```

```python
%%sql
-- check that it worked
SELECT *
FROM fires
LIMIT 5
```

```python
%%sql
--For each year, find the 10th percentile day. 
UPDATE fires AS fires 
    -- command to fill the column with DOY that lands on 10th% in order. 
    SET tenth_perctl_doy = sub.DISCOVERY_DOY
    FROM 
        -- query connect DOY with 10th percentile order/sequence
        (SELECT f.FIRE_YEAR, DISCOVERY_DOY, tenth_perctl_ct
        FROM fires AS f
        LEFT JOIN 
            -- query to identify 10th percentile in order/sequence
            (SELECT FIRE_YEAR, (MAX(count_this_year)/10) AS tenth_perctl_ct
            FROM fires
            GROUP BY FIRE_YEAR) AS g
        ON f.FIRE_YEAR = g.FIRE_YEAR
        WHERE count_this_year = tenth_perctl_ct) AS sub
    -- set so that all fires in a year will have the same DOY value in tenth_perctl_doy col 
    WHERE fires.FIRE_YEAR = sub.FIRE_YEAR
```

```python
%%sql
-- check to see that this worked. 
SELECT * 
FROM fires 
GROUP BY FIRE_YEAR
```

```python
%%sql
-- That worked. Now work toward filling in the col for the doy of each year's first big fire
SELECT FIRE_YEAR, MIN(DISCOVERY_DOY)
FROM fires
WHERE (FIRE_SIZE_CLASS IS "E") OR (FIRE_SIZE_CLASS IS "F")
GROUP BY FIRE_YEAR
```

"First fire of size e or f" might actually just not be the most useful column value to be working with. I could find the first fire of whatever size is the biggest that fires got that year, but if the biggest fire happened toward the end of the season, it wouldn't end up being a very useful metric in measuring season lenght. When I move into cleaning up and deleting columns, just delete the column I created named first_big_doy

Move into filling "peak doy" with the mean rather than 10th percentile day. Follow the same process as above. 

```python
%%sql
UPDATE fires AS fires 
    -- Filling in peak_doy instead of tenth_perctl_doy
    SET peak_doy = sub.DISCOVERY_DOY
    FROM 
        (SELECT f.FIRE_YEAR, DISCOVERY_DOY, median_ct
        FROM fires AS f
        LEFT JOIN 
            -- Here, divide by 2 instead of 10 to get the median DOY
            (SELECT FIRE_YEAR, (MAX(count_this_year)/2) AS median_ct
            FROM fires
            GROUP BY FIRE_YEAR) AS g
        ON f.FIRE_YEAR = g.FIRE_YEAR
        WHERE count_this_year = median_ct) AS sub
    WHERE fires.FIRE_YEAR = sub.FIRE_YEAR
```

```python
%%sql
-- Check to see if it worked 
SELECT * 
FROM fires 
GROUP BY FIRE_YEAR
```

Now I've got a couple options for "fire season" start dates. Let's begin with using the tenth_perctl_doy as day 1 of fire season and day 365 as the last day of fire season for each year, and add a column for fire season length. Once that is created, I can visualize over time to see if there appears to be any trend in the lenghts of fire seasons year after year. 

```python
%%sql
ALTER TABLE fires    
    ADD COLUMN season_length INTEGER
```

```python
%%sql
-- Now fill the column
UPDATE fires as f
SET season_length = sub.length
FROM 
    (SELECT FIRE_YEAR, (365-tenth_perctl_doy) AS length
    FROM fires) as sub
WHERE f.FIRE_YEAR = sub.FIRE_YEAR
```

```python
%%sql
-- Check to see if it worked 
SELECT * 
FROM fires 
GROUP BY FIRE_YEAR
```

```python
# Prepare to visualize by converting to dataframe
con = sqlite3.connect('eldorado.sqlite')
season_length_df = pd.read_sql('SELECT FIRE_YEAR, season_length FROM fires', con)
season_length_df
```

```python
sns.lineplot(data=season_length_df, x='FIRE_YEAR', y='season_length')
```

Now work to see if there is a pattern such that it would be feasible to predict future season lengths. Start by adding a simple trend line. 

```python
import numpy as np
import matplotlib.pyplot as plt

# calculate equation for trendline
z = np.polyfit(season_length_df.FIRE_YEAR, season_length_df.season_length, 1)
p = np.poly1d(z)

# Add some years into the future
trend_x = list(set(season_length_df.FIRE_YEAR)) + [2021, 2022, 2023, 2024, 2025, 2026, 2027]

# Plot the actual data and the trendline
sns.lineplot(data=season_length_df, x='FIRE_YEAR', y='season_length')
sns.lineplot(x=trend_x, y=p(trend_x))
```
